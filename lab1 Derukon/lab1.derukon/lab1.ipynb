### Контрольні запитання

__1__ _У чому відмінність інтерпретатора від компілятора? Python - компілятор чи інтерпретатор?_

_Основна відмінність між інтерпретатором і компілятором полягає в способі перетворення вихідного коду в машинний код. Компілятор перетворює весь вихідний код програми в машинний код одним разом перед запуском програми, тоді як інтерпретатор поетапно перекладає і виконує вихідний код рядок за рядком під час роботи програми. Python використовує інтерпретатор._

__2__ _Що таке IDE PyCharm?_

_PyCharm - це інтегроване середовище розробки (IDE) для мови програмування Python. Воно надає розширені можливості для редагування коду, автоматичного завершення, налагодження програм, керування проектами та багато іншого. PyCharm є популярним інструментом серед розробників Python._

__3__ _Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ svn та git?_

_Система контролю версій (СКВ) призначена для відстеження змін у файловій системі проекту та керування версіями файлів. Вона дозволяє зберігати історію змін, спільно працювати над проектами, відновлювати попередні версії файлів і вирішувати конфлікти змін. Принципова відмінність між СКВ svn (Subversion) і git полягає в їх архітектурі та способі збереження історії змін. Git є розподіленою СКВ, що означає, що кожен клон репозиторію містить повну копію історії змін, тоді як Subversion працює з централізованим репозиторієм, де зберігається історія змін._

__4__ _Що відбувається при виконанні команди `git add .`?_

_При виконанні команди `git add .`, Git додає всі модифіковані та нові файли з поточної директорії до стадії (stage), підготовляючи їх для коміту. Крапка в команді `.`, означає поточну директорію._

__5__ _Що відбувається при виконанні команди `git commit`?_

_При виконанні команди `gitcommit`, Git створює новий коміт, що представляє собою логічну одиницю змін. Коміт включає в себе зміни, які були додані до стадії, разом з повідомленням коміту, що пояснює зміст зроблених змін._

__6__ _Якою git-командою виконується розміщення локальної копії git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?_

_Для розміщення локальної копії git-репозиторію проекту на віддаленому сервері GitHub використовується команда `git push`. Вона надсилає всі локальні коміти на віддалений репозиторій, оновлюючи його зміст._
